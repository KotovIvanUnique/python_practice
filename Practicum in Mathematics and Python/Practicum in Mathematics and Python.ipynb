Способы округления
int(x) - Округляет число в сторону нуля. Это стандартная функция, для ее использования не нужно подключать модуль math.
round(x) - Округляет число до ближайшего целого. Если дробная часть числа равна 0.5, то число округляется до ближайшего четного числа.
round(x, n) - Округляет число x до n знаков после точки. Это стандартная функция, для ее использования не нужно подключать модуль math.
math.floor(x) - Округляет число вниз («пол»), при этом floor(1.5) == 1, floor(-1.5) == -2 Требует импорта math.
math.ceil(x) - Округляет число вверх («потолок»), при этом ceil(1.5) == 2, ceil(-1.5) == -1  Требует импорта math.
abs(x) - Модуль (абсолютная величина). Это — стандартная функция.
numpy.around(List, decimals=n) - Округляет все числа из numpy массива или списка List до n знаков после запятой. Требует импорта модуля numpy. Аналог round для списков/массивов/матриц.

```python
import re

def caesar(text,s):
    result = ""
 
    # traverse text
    text = re.sub(r'[\W_]+', '', text)
    for i in range(len(text)):
        char = text[i]
 
        # Encrypt uppercase characters
        if (char.isupper()):
            result += chr((ord(char) + s-65) % 26 + 65)
 
        # Encrypt lowercase characters
        else:
            result += chr((ord(char) + s - 97) % 26 + 97)
 
    return result.upper()
```

```python
def jarriquez_encryption(text, key,alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ',reverse=False):
    zcaesar = ''    
    new_text = ''
    
    for x in text.upper():
        if x in alphabet:
            new_text += x
    
    keys = (str(key)*math.ceil(len(new_text)/len(str(key))))[:len(new_text)]
    
    if reverse:
        for i in range(len(new_text)):
                zcaesar += alphabet[(alphabet.find(new_text[i]) - int(keys[i])) % len(alphabet)]
    else:
        for i in range(len(new_text)):
                zcaesar += alphabet[(alphabet.find(new_text[i]) + int(keys[i])) % len(alphabet)]
    return zcaesar
```

```python
import random
random.seed(42)

def disc_generator(alphabet):
    alphabet_list = list(alphabet)
    random.shuffle(alphabet_list)
    return ''.join(alphabet_list)

discs = []
for i in range(n):
    discs.append(disc_generator(clear_alphabet))

def jefferson_encryption(text, discs, step, reverse=False):
    zcaesar = ''
    d = 0
    for x in text.upper():
        if x in clear_alphabet:
            alfabet = discs[d]
            if reverse:
                zcaesar += alfabet[(alfabet.find(x) - step) % len(alfabet)]
            else:
                zcaesar += alfabet[(alfabet.find(x) + step) % len(alfabet)]
            if d < len(discs)-1:
                d += 1
            else:
                d = 0
    return zcaesar
```

```python
import re
import random

def mimic_dict(string):
    s = re.split(' |\t|\n', string)
    d = {}
    for i in range(-1,len(s)-1):
        if i == -1:
            d['']=[s[i+1]]
        elif s[i] in d:
            d[s[i]] = d.get(s[i])+[s[i+1]]
        elif s[i+1]:
            d[s[i]]=[s[i+1]]
    return d
```

```python
import random
random.seed(42)

def print_mimic(mimic_dict, word):
    """Given mimic dict and start word, return 200 random words."""
    
    c = 0
    next_word = word
    result = word
    
    while c < 200:
        try:
            last_word = random.choice(mimic_dict.get(next_word))
        except:
            last_word = random.choice(mimic_dict.get(''))
        result += ' '+last_word
        next_word = last_word
        c += 1
        
    return result
```

```python
import pandas as pd
import numpy as np

s = list(map(int,input().split()))
df = pd.DataFrame(s)
modes = np.array(df.mode())
mean = round(float(df.mean()[0]),1)
median = round(float(df.median()[0]),1)

print('Среднее: '+str(mean))
print('Медиана: '+str(median))

if len(modes) == 1:
    j = modes[0][0]
    j_float = float(j)
    print('Мода: '+str(j),end=' ')
else:
    print('Моды:',end=' ')    
    for i in modes:
        j_float = float(i[0])
        print(i[0],end=' ')
        
print('')
if True in [x for x in modes > mean]:
    print('Мо>Ср')
if True in [x for x in modes == mean]:
    print('Мо=Ср')
if True in [x for x in modes < mean]:
    print('Мо<Ср')
if True in [x for x in modes > median]:
    print('Мо>Ме')
if True in [x for x in modes == median]:
    print('Мо=Ме')
if True in [x for x in modes < median]:
    print('Мо<Ме') 
if mean > median:
    print('Ме<Ср')
elif mean == median:
    print('Ме=Ср')
else:
    print('Ме>Ср')                
```
